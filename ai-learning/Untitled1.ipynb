```python
import pandas as pd
df = pd.DataFrame({
    'name':   ['Alice', 'Bob', 'Carol', 'Dave'],
    'age':    [1, 30, 35, 28],
    'score':    [1, 30, 35, 28],
    'salary': [70000, 85000, 90000, 75000],
    'dept':   ['Eng', 'Sales', 'Eng', 'HR']
})
```

```python
df['age_group'] = df['age'].apply(lambda x:
    'young'  if x < 30 else
    'mid'    if x < 45 else
    'senior'
)
df
```

```python
df['value_score'] = df.apply(
    lambda row: row['salary'] / row['age'] * row['score'],
    axis=1    # axis=1 means "apply across columns for each row"
)
df
```
