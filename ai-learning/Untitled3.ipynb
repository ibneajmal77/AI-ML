```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
epochs = np.arange(1, 21)
train_loss = np.exp(-epochs * 0.2) + np.random.randn(20) * 0.02
val_loss   = np.exp(-epochs * 0.15) + np.random.randn(20) * 0.03 + 0.1

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(epochs, train_loss, label='Training Loss',   color='blue',   linewidth=2)
ax.plot(epochs, val_loss,   label='Validation Loss', color='orange', linewidth=2, linestyle='--')
ax.set_title('Training vs Validation Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
ax.legend()
ax.grid(True, alpha=1)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Set style once at top of notebook
sns.set_theme(style='whitegrid', palette='muted')

# seaborn understands DataFrames directly — no need to extract columns manually
df = pd.DataFrame({
    'salary': np.random.randint(50000, 120000, 100),
    'age':    np.random.randint(22, 55, 100),
    'dept':   np.random.choice(['Eng', 'Sales', 'HR'], 100),
    'score':  np.random.uniform(60, 100, 100)
})
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Basic
sns.histplot(df['salary'], bins=30, kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Salary Distribution')
# Grouped by category — overlapping distributions
sns.histplot(df, x='salary', hue='dept', kde=True, ax=axes[1], alpha=0.5)
axes[1].set_title('Salary Distribution by Department')

plt.tight_layout()
plt.show()
```

```python
# Create a more realistic dataset
np.random.seed(42)
n = 200
df_corr = pd.DataFrame({
    'age':        np.random.randint(22, 60, n),
    'salary':     np.random.randint(40000, 150000, n),
    'experience': np.random.randint(1, 35, n),
    'score':      np.random.uniform(50, 100, n),
    'bonus':      np.random.randint(1000, 20000, n),
})
df_corr['salary']     = df_corr['salary'] + df_corr['experience'] * 2000  # introduce correlation
df_corr['bonus']      = df_corr['salary'] * 0.1 + np.random.randn(n) * 500

corr_matrix = df_corr.corr()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,          # show the number in each cell
    fmt='.2f',           # 2 decimal places
    cmap='coolwarm',     # red=positive, blue=negative correlation
    center=0,            # 0 correlation = white
    square=True,
    linewidths=0.5,
    ax=ax
)
ax.set_title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
