```python
import pandas as pd

df = pd.read_csv("sample_superstore_like_dataset.csv")

print(df.head())
```

```python
import matplotlib.pyplot as plt

# group data
category_sales = df.groupby("Category")["Sales"].sum()

# plot
category_sales.plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values)

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.grid(axis='y')

plt.show()
```
